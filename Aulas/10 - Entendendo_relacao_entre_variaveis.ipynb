```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins_numeric = penguins.drop(columns = ['species', 'island', 'sex'])
```

```python
penguins_numeric.corr()
```

```python
for col in penguins.drop(columns = ['species', 'island', 'sex']):
  sns.barplot(x = 'species', y = col, data = penguins)
  plt.show()
```

```python
sns.scatterplot(x = 'body_mass_g', y = 'flipper_length_mm', data = penguins, hue = 'species')
```

```python
sns.scatterplot(x = 'body_mass_g', y = 'bill_length_mm', data = penguins, hue = 'species')
```

```python
sns.pairplot(data=penguins, hue='species')
```
