# Importar Pandas e Numpy

```python
import pandas as pd
import numpy as np
```

# Importar base houses

```python
houses_sp = pd.read_csv('../dados/houses_sp.csv')
houses_sp.head(10)
```

# Verificar a distribuição dos imoveis no dataset

```python
houses_sp['city'].value_counts()
```

```python
houses_sp.info()
```

### Remover a coluna city por que não é interessante uma vez que só tem dados de São Paulo, esse dado é chamado de dados constante.

```python
houses_sp = houses_sp.drop(columns=['city'])
houses_sp.head(10)
```

## Verificar dados NAN

```python
houses_sp['rooms']
```

## Visualizando informações e quantidade de NaN

```python
houses_sp.info()
```

### Tratar dados NAN usando mediana

```python
houses_sp['rooms'].median()
```

```python
houses_sp['rooms'] = houses_sp['rooms'].fillna(houses_sp['rooms'].median())
houses_sp.info()
```

### Verificar dados digitados erroneamentes

```python
houses_sp.head(10)
```

```python
houses_sp['floor'].value_counts()
```
