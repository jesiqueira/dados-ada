```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data set iris
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
# visualizar com histograma
col = 'sepal_length'
sns.histplot(data=iris, x=col, kde=True).set_title('Distribuição da variável' + col)
```

```python
for col in iris.drop(columns='species').columns:
    sns.histplot(data=iris, x=col, kde=True).set_title('Distribuição da variável ' + col)
    plt.show()
```

```python
for col in iris.drop(columns='species').columns:
    sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title('Distribuição da variável ' + col)
    plt.show()
```

```python
for col in iris.drop(columns='species').columns:
    sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y= col,ci=90)
    plt.show()
```
